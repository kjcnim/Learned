```
import tensorflow as tf
import numpy as np

#training set 
x_train = [1,2,3]
y_train = [1,2,3]

#keras의 다차원 계층 모델인 Sequential를 레이어 만든다.
tf.model = tf.keras.Sequential() 

#입력이 1차원, 출력이 1차인임을 뜻한다. Dense는 layer의 종류
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = 1)) 

#Optimizer - Stochastic gradient descent : 확률적 경사 하강법
sgd = tf.keras.optimizers.SGD(lr=0.1) 

#cost/loss function
#loss를 mean_squared_error 방식을 사용한다는 의미
tf.model.compile(loss = 'mse', optimizer = sgd)

#print summary of the model to the terminal
tf.model.summary()

#fit()executes training
#tensor flow 1과 다르게 세션을 만들어서 돌릴 필요가 없다.
tf.model.fit(x_train, y_train, epochs = 200)

#predict() returns predicted value
y_predict = tf.model.predict(np.array([5,4,7]))
print(y_predict)
```
