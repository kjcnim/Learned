```
import tensorflow as tf

x_data = [[1,2],
         [2,3],
         [3,1],
         [4,3],
         [5,3],
         [6,2]]
y_data = [[0],
         [0],
         [0],
         [1],
         [1],
         [1]]

tf.model = tf.keras.Sequential()

#input이 2차원 출력이 1차원이다.
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = 2))

#sigmoid 함수를 사용하여 0과 1 사이의 값으로 나타낸다.
#sigmoid 함수를 사용하는 이유에 대하여...
#이전의 regression은 hypothesis, cost, gradient descent 3가지를 통해 숫자를 예측하는 것이었다면
#binary classification은 둘 중 한개를 정해진 카테고리 중 하나를 선택하는 것이다.(spam or ham, show or hide... etc)
#먼저, Logistic regression은 이진 분류(binary classification)문제를 해결하기 위한 모델이다.
#Binary classification이란 단어 그대로 데이터를 0 혹은 1로 분류하는 것을 말한다.
#이를 직관적으로 할 수 있는 것이 선을 긋는 방식이다. 즉, Linear Regression 분석방식.
#하지만 일반적인 H(x)값으로 Binary classification을 하기란 어렵다. 왜냐하면 값이 0보다 작을수도 있고 1보다 클 수도 있기에.
#따라서 이 상황에서 고려해야할 것은 H(x)값인데,
#H(x)값이 0 혹은 1 사이의 값이어야한다.
#이를 가능하게 하는 것이 바로 'sigmoid'함수이다. 그래서 이를 사용하는 것이다.
#즉, H(x) 수식이 sigmoid 함수꼴이 된다.
#sigmoid를 적용한 공식이 Logistic Hypothesis이다.
#이를 이용한 cost 식도 다르게 되는데, 그 이유는 sigmoid 그래프의 특성 때문이다. (이전의 regression cost 식과는 다른 식이된다.)
#마지막으로 구해지는 cost를 minimize 해야한다. 
#이 방식은 이전과 동일하게 gradient descent 방식 그대로 적용한다.


tf.model.add(tf.keras.layers.Activation('sigmoid'))

#loss = 'mes'보다 binary_crossentropy로 하는게 더 좋은 성능을 보인다.
#accuracy metrics를 추가함으로써 더 좋은 트레이닝 정확도를 얻을 수 있다.
tf.model.compile(loss = 'binary_crossentropy', optimizer = tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])

tf.model.summary()

#epoch를 5000으로하면 Accuracy가 1이 됨
history = tf.model.fit(x_data, y_data, epochs = 500)

print("Accuracy :",history.history['accuracy'][-1])
```
